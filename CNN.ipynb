```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
data = pd.read_csv('EEG.machinelearing_data_BRMH.csv')
data.head()
```

```python
data = data.drop(columns=['Unnamed: 122'])
```

```python
data.drop(["sex", "no.", "age", "eeg.date","education", "IQ"], axis=1, inplace =True)
data.head()
```

```python
data.rename(columns={"main.disorder":"main_disorder"}, inplace = True)
data.rename(columns={"specific.disorder":"specific_disorder"}, inplace = True)
data.head()
```

```python
data.drop(["main_disorder"], axis=1, inplace =True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
classes_to_keep = ['Alcohol use disorder', 'Behavioral addiction disorder', 'Depressive disorder', 'Healthy control', 'Schizophrenia']
data_filtered = data[data['specific_disorder'].isin(classes_to_keep)]

# Then split into features and labels
x_filtered = data_filtered.drop(columns=['specific_disorder']).values
y_filtered = data_filtered['specific_disorder'].values.ravel()

le = LabelEncoder()
y_encoded = le.fit_transform(y_filtered)
Y = y_encoded
X = x_filtered
```

```python
x_train, x_temp, y_train, y_temp = train_test_split(
    X, Y, test_size=0.3, stratify=Y, random_state=420)

x_val, x_test, y_val, y_test = train_test_split(
    x_temp, y_temp, test_size=0.5, stratify=y_temp, random_state=420)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_val = scaler.transform(x_val)
x_test = scaler.transform(x_test)
```

```python
x_train_t, x_temp_t, y_train_t, y_temp_t = train_test_split(
    X, Y, test_size=0.3, stratify=Y, random_state=420)
```

```python
from tensorflow.keras.utils import to_categorical

# One-hot encode
y_train = to_categorical(y_train, num_classes=5)
y_test = to_categorical(y_test, num_classes=5)
y_val = to_categorical(y_val, num_classes=5)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, LeakyReLU
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.utils import class_weight

# Reshape input for Conv1D 
x_train_cnn = x_train.reshape((x_train.shape[0], x_train.shape[1], 1))
x_test_cnn = x_test.reshape((x_test.shape[0], x_test.shape[1], 1))
x_val_cnn = x_val.reshape((x_val.shape[0], x_val.shape[1], 1))

# Build the model
model = Sequential([
    Conv1D(16, kernel_size=5, input_shape=(1140, 1)),
    LeakyReLU(alpha=0.1),
    MaxPooling1D(pool_size=2),

    Conv1D(16, kernel_size=5),
    LeakyReLU(alpha=0.1),
    MaxPooling1D(pool_size=2),

    Flatten(),
    Dropout(0.4),
    Dense(16, kernel_regularizer=regularizers.l2(0.001)),
    LeakyReLU(alpha=0.1),
    
    Dense(5, activation='softmax')  # 5 output classes
])

# Compile
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Add early stop
early_stop = EarlyStopping(
    monitor='val_loss',      
    patience=5,              
    restore_best_weights=True  
)

# Class weights
class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train_t),
    y=y_train_t
)

class_weights_dict = dict(enumerate(class_weights))


# Train
history = model.fit(x_train_cnn, y_train,
                    epochs=10,
                    batch_size=16,
                    validation_data=(x_val_cnn, y_val),
                    class_weight=class_weights_dict,
                    callbacks=[early_stop])
```

```python
# Evaluate
loss, acc = model.evaluate(x_test_cnn, y_test)
print(f"Test accuracy: {acc:.2f}")

# Predict class indices
y_pred = model.predict(x_test_cnn)
y_pred_classes = y_pred.argmax(axis=1)
```
